## ‚úÖ **1. NDWI Formula and Band Rationale**

**Formula:**

$$
\text{NDWI} = \frac{Green - NIR}{Green + NIR}
$$

**For Sentinel-2:**

* Green = **B3**
* NIR = **B8**

$$
\text{NDWI} = \frac{B3 - B8}{B3 + B8}
$$

**Rationale:**

* **Green (B3)** reflects well from vegetation and water, helping to distinguish surfaces.
* **NIR (B8)** is strongly absorbed by water and highly reflected by vegetation and soil, making it effective for separating water from land.

---

## ‚úÖ **2. NDWI vs. MNDWI**

### **NDWI (Normalized Difference Water Index):**

$$
\text{NDWI} = \frac{Green - NIR}{Green + NIR}
$$

### **MNDWI (Modified NDWI):**

$$
\text{MNDWI} = \frac{Green - SWIR}{Green + SWIR}
$$

**For Sentinel-2:**

* Green = B3
* SWIR = **B11** (or sometimes B12)

**MNDWI =**

$$
\frac{B3 - B11}{B3 + B11}
$$

**Difference and Situations for MNDWI Use:**

* **Key Difference:** The primary difference is the substitution of the NIR band with a **SWIR band** in MNDWI.
* **Why SWIR?** Water absorbs SWIR radiation even more strongly than NIR radiation. Soil, vegetation, and especially built-up land features tend to have higher reflectance in the SWIR region compared to water.
* **Advantages of MNDWI:**
    * **Enhanced Water Feature Extraction:** MNDWI often provides better discrimination of water bodies from other features, especially in areas where NDWI might struggle.
    * **Reduced Noise from Built-up Areas:** Built-up land and artificial structures can sometimes have NIR reflectance similar to water (or cause mixed pixel effects), leading to their misclassification as water by NDWI. These features generally have higher SWIR reflectance than water, so MNDWI can suppress signals from built-up areas more effectively, resulting in cleaner water masks.
    * **Improved Performance with Turbid Water:** While both indices work for clear water, MNDWI can sometimes offer better delineation of turbid water.



### **When to Use MNDWI:**

* In **urban areas**, where built-up surfaces reflect NIR similar to water, causing NDWI to confuse water with man-made objects.
* MNDWI uses **SWIR**, which is **highly absorbed by water** and **reflected by buildings**, improving discrimination.

---

## ‚úÖ **3. Why is NIR Effective for Water Detection?**

* Water **absorbs** most NIR radiation ‚Üí appears **dark** in NIR images.
* Vegetation and soil **reflect** NIR ‚Üí appear **bright**.
* This spectral contrast makes it easy to isolate water.

## ‚úÖ **4. Calculating Indices & Analysis**

**Given Pixel Values:**

* B3 (Green): 1
* B8 (NIR): 9
* B11 (SWIR): 12
* B4 (Red): 9

---

### **NDWI:**

$$
\frac{B3 - B8}{B3 + B8} = \frac{1 - 9}{1 + 9} = \frac{-8}{10} = \boxed{-0.8}
$$

---

### **MNDWI:**

$$
\frac{B3 - B11}{B3 + B11} = \frac{1 - 12}{1 + 12} = \frac{-11}{13} ‚âà \boxed{-0.846}
$$

---

### **NDVI (Normalized Difference Vegetation Index):**

$$
\text{NDVI} = \frac{NIR - Red}{NIR + Red} = \frac{B8 - B4}{B8 + B4} = \frac{9 - 9}{9 + 9} = \frac{0}{18} = \boxed{0}
$$

---

## ‚úÖ **Analysis:**

* **NDWI = -0.8** ‚Üí Strong **non-water** indication.
* **MNDWI = -0.846** ‚Üí Even stronger confirmation of **non-water** (likely urban/built-up or dry soil).
* **NDVI = 0** ‚Üí Indicates **bare soil or built-up areas**, not vegetation.
* **NDVI = 0.00:**
    NDVI values range from -1 to +1.
    * Healthy, dense vegetation usually has high positive NDVI values (e.g., 0.6 to 0.9).
    * Sparse vegetation or stressed vegetation has lower positive values.
    * Bare soil, rock, and sand typically have NDVI values close to zero or slightly positive (e.g., 0.1 to 0.2).
    * Snow, ice, and clouds often have NDVI values around zero or slightly negative.
    * Water bodies usually have negative NDVI values.
    An NDVI of 0.00 means the Red reflectance (9) is equal to the NIR reflectance (9). This generally points to a **non-vegetated surface**. This could be bare soil, rock, sand, or an artificial surface. It rules out healthy vegetation.
